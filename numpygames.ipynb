```python
#import relevant libraries and show things on board
import numpy as np
import matplotlib.pyplot as plt
def displayboard(board):
    plt.imshow(board)
    plt.show()
```

```python
def conway(s, p):
    if p==0:
        a=np.zeros((s,s),dtype = np.int)
    elif p==1 :
        a=np.ones((s,s),dtype = np.int)
    else :
         a= np.random.choice(2, s*s,p=[1-p,p]).reshape(s, s) 
    print('showing original board')
    displayboard(a)
    return a
```

```python
def advance(b,t) :
    new=b.copy()
    N=int(b.shape[0])
    for m in range(1,t+1):
        b2 = b.copy()
       # print('showing step {} of advance'.format(t))
       # print('showing step {} of advance'.format(m))
        for i in range(N):
            for j in range(N):
                sum = int((b[i, (j-1)%N] + b[i, (j+1)%N] +
                    b[(i-1)%N, j] + b[(i+1)%N, j] +
                    b[(i-1)%N, (j-1)%N] + b[(i-1)%N, (j+1)%N] +
                    b[(i+1)%N, (j-1)%N] + b[(i+1)%N, (j+1)%N]))
                if b[i, j] == 1:
                    if sum < 2 or sum > 3:
                        b2[i, j] = 0
                else:
                    if sum == 3:
                        b2[i, j] = 1
        print('advance of {} steps'.format(m))
        displayboard(b2)
        b = b2
```

```python
# s =6, p = 0.1, t = 3
b = conway(6, 0.5)
advance(b, 3)
```

```python
# s =1000, p = 0.1, t = 5
b = conway(1000, 0.1)
advance(b, 5)
```
